```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import datetime
```

```python
df = pd.read_csv('../Resources/Beach_Lab_Data.csv')
```

```python
df.sort_values('DNA Sample Timestamp',ascending=False)
```

```python
df['Beach'].value_counts()
```

```python
only_ohio = df.loc[df["Beach"] == "Ohio Street", :]
```

```python
only_ohio
```

```python
only_ohio_clean = only_ohio[only_ohio['DNA Reading Mean'].notna()]
```

```python
only_ohio_clean = only_ohio_clean.drop(columns=['Culture Test ID','Culture Sample 1 Timestamp','Culture Sample 1 Reading','Culture Sample 2 Reading','Culture Reading Mean','Culture Note','Culture Sample Interval','Culture Sample 2 Timestamp'])
```

```python
only_ohio_clean['Date'] = pd.to_datetime(only_ohio_clean['DNA Sample Timestamp'], errors = 'coerce')
```

```python
only_ohio_clean['Sample_Date'] = pd.to_datetime(only_ohio_clean['Date']).dt.to_period('D')
```

```python
only_ohio_clean = only_ohio_clean[only_ohio_clean['Sample_Date'].notna()]
```

```python
only_ohio_clean.sort_values('Sample_Date',ascending=False)
```

```python
only_ohio_clean.groupby('Sample_Date').mean()
```

```python
water_quality = pd.read_csv('../Resources/Beach_Water_Quality.csv')
water_quality.head()
```

```python
water_only_ohio = water_quality.loc[water_quality["Beach Name"] == "Ohio Street Beach", :]
water_only_ohio
```
