```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import datetime
```

```python
df = pd.read_csv('../Resources/Beach_Lab_Data.csv')
```

```python
df.sort_values('DNA Sample Timestamp',ascending=False)
```

```python
df['Beach'].value_counts()
```

```python
only_calumet = df.loc[df["Beach"] == "Calumet", :]
```

```python
only_calumet_clean = only_calumet[only_calumet['DNA Reading Mean'].notna()]
```

```python
only_calumet_clean = only_calumet_clean.drop(columns=['Culture Test ID','Culture Sample 1 Timestamp','Culture Sample 1 Reading','Culture Sample 2 Reading','Culture Reading Mean','Culture Note','Culture Sample Interval','Culture Sample 2 Timestamp'])
```

```python
only_calumet_clean['Sample_Month_Year'] = pd.to_datetime(only_calumet_clean['DNA Sample Timestamp']).dt.to_period('M')
```

```python
only_calumet_clean.sort_values('Sample_Month_Year',ascending=False)
```
