```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import datetime
```

```python
df = pd.read_csv('../Resources/Beach_Lab_Data.csv')
```

```python
df['Beach'].value_counts()
```

```python
only_12th_street = df.loc[df["Beach"] == "12th Street", :]
```

```python
only_12th_street_clean = only_12th_street[only_12th_street['Culture Reading Mean'].notna()]
```

```python
only_12th_street_clean = only_12th_street_clean.drop(columns=['DNA Test ID','DNA Sample Timestamp','DNA Sample 1 Reading','DNA Sample 2 Reading','DNA Reading Mean'])
```

```python
only_12th_street_clean['Sample1_Month_Year'] = pd.to_datetime(only_12th_street_clean['Culture Sample 1 Timestamp']).dt.to_period('M')
```

```python
only_12th_street_clean['Sample2_Month_Year'] = pd.to_datetime(only_12th_street_clean['Culture Sample 2 Timestamp']).dt.to_period('M')
```

```python
only_12th_street_clean.sort_values('Sample1_Month_Year',ascending=False)
```
